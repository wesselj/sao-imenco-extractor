### Create Time series

```python
from cognite.client.data_classes.time_series import TimeSeries

def get_imenco_name_suffix(name: str):
    return name[len("Imenco."):]

imenco_camera_sensors_and_units = [
    ("Temperature", "deg C"),
    ("Depth", "m"),
    ("OxygenLevel", "mg/L"),
    ("OxygenLevelPercent", "%"),
    ("internalHeading", "deg"),
    ("topCameraHeading", "deg"),
    ("bottomCameraHeading", "deg")
]

imenco_ts_list = []
for cam_sensor, unit in imenco_camera_sensors_and_units:
    for cam_name_ext_id, cam_asset_id in imenco_ext_id_id:
        cam_name = get_imenco_name_suffix(cam_name_ext_id)
        imenco_ts_list.append(TimeSeries(
            external_id=f"imenco.ocean_farm_1.{cam_name}.{cam_sensor}",
            name=f"imenco.ocean_farm_1.{cam_name}.{cam_sensor}",
            unit=unit,
            is_step=False,
            is_string=False,
            asset_id=cam_asset_id,
            data_set_id=imenco_ds.id,
            description=f"Imenco feeding camera '{cam_sensor}' data from camera: {cam_name} on Ocean Farm 1. See: https://imenco.no/product/pixelite-insight",
            metadata = {
                "src_data_type": "measurements",
                "src_equipment": get_imenco_name_suffix(cam_name_ext_id),
                "src_equipment_sensor": cam_sensor,
                "src_equipment_data_unit": unit,
                "src_system": "imenco",
                "src_data_externalId_namespace1": "imenco",
                "src_data_externalId_namespace2": "ocean_farm_1",
                "ops_extractor_name": "rest:imenco:ocean_farm_1:ts",
                "ops_extractor_pipeline_id": 4741,
                "supplier": "Imenco",
                "location": "Habranden",
                "farm": "Ocean Farm 1",
                "IP": Imenco_camera_IP_map[cam_name]
            })
        )

def show():
    [print(ts) for ts in imenco_ts_list]
# show()
#client.time_series.create(imenco_ts_list) #DONE
```
